```python
import pandas as pd
```

```python
df=pd.read_csv("titanic.csv")
df
```

#Deleting the unnecessary Columns


```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
```

```python
df
```

# Making a new Variable where Survived column is stored

```python
inputs = df.drop('Survived',axis='columns')
target = df.Survived
```

```python
inputs
```

```python
target
```

# Mappping the values of Gender as 1 or 2

```python
inputs.Sex = inputs.Sex.map({'male': 1, 'female': 2})
```

```python
inputs
```

```python
inputs.head(10)
```

#We saw that the data of Sex Columns is conataing NaN values at 5th so it can be assured that other row can also have NaN values so removing that values with the mean of the ages


```python
inputs.Age = inputs.Age.fillna(inputs.Age.mean())
inputs.head()
```

#Spliting the dataset into Train and Test

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs,target,test_size=0.2)
```

```python
len(X_train)
```

```python
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```
